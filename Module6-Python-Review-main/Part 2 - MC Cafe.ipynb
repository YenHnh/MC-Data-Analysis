# Part 2 - MC Cafe

Your instructors love coffee so much that they decided to open a cafe! Of course, we will need the help of our students to develop the ordering app that the customers can use.

On the cafe menu, there are three categories for the items we sell - coffee, tea, and pastries.

<b>Coffee:</b>

$$ Espresso - $1.50 $$
$$Americano - $2.00$$
$$Caffe latte - $2.25$$
$$Mocha - $2.25$$
$$Cappucino - $2.50$$
$$Macchiato - $2.50 $$


<b>Tea</b>:

$$White tea - $1.50$$
$$Green tea - $2.00$$
$$Oolong tea - $2.25$$
$$Black tea - $2.25$$
$$Earl Grey - $2.50$$
$$Matcha - $2.50$$


<b>Pastries:</b>

$$Croissant - $5.50$$
$$Cinnamon roll - $4.50$$
$$Banana bread - $4.25$$
$$Pumpkin bread - $3.75$$
$$Sticky bun - $4.25$$
$$Apple turnover - $3.75$$

When your program first starts, you must greet the customers by saying, "Welcome to MC Cafe!" Then ask the customer if they would like to see the menu. If yes, then display the menu items and prices.

Then ask the user if they would like to order. If yes, then prompt the customer to enter an item and the quantity that they want. After they input an item, then ask them if they would like to order another item.

Once your customers are finished with the order input, then display an order receipt that lists the items that they ordered and the quantity. Then at the bottom of the receipt, display the subtotal (the amount of the item before tax), the amount in sales tax (our business is in Maryland, so state tax is 6%), and finally their grand total (subtotal plus sales tax).

At the end of their transaction, display a message to our customers saying, "Thank you for visiting MC Cafe!"

BONUS: Add a feature that tells a customer when an item they attempted to order is not on the menu.

```python
print("\tWelcome to MC Cafe!")
print("")

list_menu = {"Espresso":1.50,"Americano":2.00,"Caffe latte":2.25,"Mocha":2.25,"Cappucino":2.50,"Macchiato":2.50,"White tea":1.50,"Green tea":2.00,"Oolong tea":2.25,"Black tea":2.25,"Earl Grey":2.50,"Matcha":2.50,"Croissant":5.50,"Cinnamon roll":4.50,"Banana bread":4.25,"Pumpkin bread":3.75,"Sticky bun":4.25,"Apple turnover":3.75}

menu = input("Would like to see the menu? (y/n)")

if menu.lower() != "n":
    print("\nCoffee:\n\nEspresso - $1.50\nAmericano - $2.00\nCaffe latte - $2.25\nMocha - $2.25\nCappucino - $2.50\nMacchiato - $2.50\n")
    print("\nTea:\n\nWhite tea - $1.50\nGreen tea - $2.00\nOolong tea - $2.25\nBlack tea - $2.25\nEarl Grey - $2.50\nMatcha - $2.50\n")
    print("\nPastries:\n\nCroissant - $5.50\nCinnamon roll - $4.50\nBanana bread - $4.25\nPumpkin bread - $3.75\nSticky bun - $4.25\nApple turnover - $3.75\n")

order = input("Would you like to order? (y/n)")

if order.lower() != "n":
    order_again = ""
    total= 0
    result = {}
    
    while order_again.lower() != "n":
        purchase = input("Enter the name of the item you want to order: ")
        
        if purchase not in list_menu.keys():
            print("That item isn't on the menu")
            
        for i in list_menu.keys():
            if purchase == i:
                quan = int(input("Enter the quality you want to order: "))
                cost = quan * list_menu[purchase]
                total += cost
                result[purchase] = {quan}
                  
        order_again = input("Would you like to order another item? (y/n) ")
        
     
    tax = 0.06 * total
    print("-"*100)
    print("\t\tReceipt\n")
    print(f"\tYou ordered: ")
    print(result)
    print(f"\tSubtotal:    {total:.2f}$")
    print(f"\tSale tax:    {tax:.2f}$")
    print(f"\tTotal:       {(total + tax):,.2f}$")
    
print("")    
print("\tThank you for visiting MC Cafe!")
```
